## 2. a Create a new column -’total_benefitted’ that is a sum of inmates benefitted through all modes.

```python
data.head()
```

```python
data['total_benefitted'] = data['No. of Inmates benefitted by Elementary Education'] + data['No. of Inmates benefitted by Adult Education'] + data['No. of Inmates benefitted by Higher Education'] + data['No. of Inmates benefitted by Computer Course']
```

```python
data.head()
```

## b. Create a new row - “totals” that is the sum of all inmates benefitted through each mode across all states

```python
totals = data.iloc[:, 2:].sum()
```

```python
totals.name = 'totals'
data = data.append(totals.transpose())
```

```python
data.tail()
```

## Plotting:

## a. Make a bar plot with each state name on the x -axis and their total benefitted inmates as their bar heights. Which state has the maximum number of beneficiaries?

```python
data = data.append(data.sum(), ignore_index=True)
```

```python
data.head()
```

```python
x = data["STATE/UT"]
y = data["total_benefitted"]
```

```python
plt.bar(x,y)
plt.setp(plt.gca().get_xticklabels(), rotation=90,horizontalalignment='right') 
plt.title("STATE/UT vs Total_inmates_benefitted")
plt.xlabel("STATE/UT")
plt.ylabel("Total_inmates_benefitted")
plt.show()
```

## b. Make a pie chart that depicts the ratio among different modes of benefits.

```python
labels = data["STATE/UT"]
values = data["total_benefitted"]
```

```python
plt.pie(values, labels = labels, autopct='%1.1f%%', shadow=True)
plt.title("STATE/UT vs Total_inmates_benefitted")
plt.show()
```

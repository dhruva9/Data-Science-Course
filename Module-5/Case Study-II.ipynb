## Compute how much total salary cost has increased from year 2011 to 2014

```python
data.head(3)
```

```python
salary_increse = data.groupby('Year')[['TotalPay']].sum()
```

```python
salary_increse
```

```python
salary_increse.plot();
```

```python
salary_increse['TotalPay']
```

```python
salary_increse.reset_index(inplace=True)
```

```python
salary_increse
```

```python
total_increse = salary_increse['TotalPay'][3] - salary_increse['TotalPay'][0]
```

```python
print("Total change from 2011 to 2014 is : ", total_increse)
```

## Which Job Title in Year 2014 has highest mean salary?

```python
data.head(3)
```

```python
jobpay_2014 = data.query('Year == 2014').groupby(['JobTitle']).agg('mean').filter(['JobTitle', 'TotalPay'])
```

```python
print(jobpay_2014.sort_values(by='TotalPay', ascending=False).head(1))
```

## Which are the top 5 common job in Year 2014 and how much do they cost SFO ?

```python
data.head(2)
```

```python
top5_in2014 = data.groupby(['Year', 'JobTitle']).agg('count').filter(['Year', 'TotalPay', 'JobTitle']).query("Year == 2014")
```

```python
print("top 5 common job in Year 2014")
top5_in2014.sort_values('TotalPay', ascending=False).head()
```

## Who was the top earning employee across all the years?

```python
data.head(3)
```

```python
top_earning = data.groupby('EmployeeName')[['TotalPay']].sum()
```

```python
print("Top earning employee across all the years")
top_earning.sort_values(by='TotalPay', ascending=False).head()
```
